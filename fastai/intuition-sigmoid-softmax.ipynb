## Sigmoid

Motivation: Squeeze the whole numeric range (x-values) to the range from zero to one (y-values).

Properties:

- monotonically nondecreasing (monoton steigend)
- differentiable
- non-linear: (very) small values are (almost) zero, (very) large values are (almost) one

### Start with an exponential function

```python
def f1(x):
    return np.exp(x)
plot_function(f1, title="e(x)", tx="x", ty="y", min=-4, max=4)
```

### Mirror the function (y-axis)

```python
def f2(x):
    return np.exp(-x)
plot_function(f2, title="e(-x)", tx="x", ty="y", min=-4, max=4)
```

### Shift by +1 to the top

```python
def f3(x):
    return 1+np.exp(-x)
plot_function(f3, title="1+e(-x)", tx="x", ty="y", min=-4, max=4)
```

### Reciprocal (1/...) to create final Sigmoid with S-curve shape and nice convergence property

This is the final Sigmoid form!

```python
def f4(x):
    return 1/(1+np.exp(-x))
plot_function(f4, title="1 / (1+e(-x))", tx="x", ty="y", min=-4, max=4)
```

### What if... we hadn't shifted by +1 in the denominator?

We would get values smaller than 1 in the denominator, which would not give a saturating behavior, but instead the function goes to infinity.

```python
def f5(x):
    return 1/(np.exp(-x))
plot_function(f5, title="1 / (e(-x))", tx="x", ty="y", min=-4, max=4)
```
